# Feature Engineering & Preprocessing

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

os.chdir(r'C:\Users\USP\Documents\Projects\house-price-project')

df = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# Keep the test IDs for submission later
test_ids = test['Id']

# Combine train and test so preprocessing is identical on both
all_data = pd.concat([df.drop('SalePrice', axis=1), test], axis=0).reset_index(drop=True)

print(f"Train: {df.shape}, Test: {test.shape}, Combined: {all_data.shape}")
```

```python
# Log-transform the target variable to reduce skewness
y = np.log1p(df['SalePrice'])

plt.figure(figsize=(8, 4))
plt.hist(y, bins=50, color='coral', edgecolor='white')
plt.title('Log-transformed SalePrice (your target variable)')
plt.xlabel('log(SalePrice)')
plt.show()
```

```python
# Two houses with huge square footage but very low prices — spotted in EDA
outlier_idx = df[(df['GrLivArea'] > 4000) & (df['SalePrice'] < 200000)].index
df = df.drop(outlier_idx)
y = np.log1p(df['SalePrice'])

print(f"Removed {len(outlier_idx)} outliers. Training rows remaining: {len(df)}")
```

```python
# These columns are missing because the feature doesn't exist (no pool, no garage, etc.)
none_cols = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu',
             'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
             'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1',
             'BsmtFinType2', 'MasVnrType', 'MSSubClass']

for col in none_cols:
    all_data[col] = all_data[col].fillna('None')

# These columns are missing because the value is unknown — fill with the most common value
mode_cols = ['MSZoning', 'Electrical', 'KitchenQual', 'Exterior1st',
             'Exterior2nd', 'SaleType', 'Functional']

for col in mode_cols:
    all_data[col] = all_data[col].fillna(all_data[col].mode()[0])

print("Categorical nulls filled.")
```

```python
# LotFrontage: street-connected frontage. Fill with median per neighborhood
# (houses in the same neighborhood tend to have similar lot sizes)
all_data['LotFrontage'] = all_data.groupby('Neighborhood')['LotFrontage']\
                          .transform(lambda x: x.fillna(x.median()))

# Garage year built: if no garage, year is 0
all_data['GarageYrBlt'] = all_data['GarageYrBlt'].fillna(0)

# Area columns: no feature = 0 square feet
zero_cols = ['GarageArea', 'GarageCars', 'BsmtFinSF1', 'BsmtFinSF2',
             'BsmtUnfSF', 'TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath', 'MasVnrArea']

for col in zero_cols:
    all_data[col] = all_data[col].fillna(0)

print(f"Remaining nulls: {all_data.isnull().sum().sum()}")
```

```python
all_data.isnull().sum()[all_data.isnull().sum() > 0] 
```

```python
all_data['Utilities'] = all_data['Utilities'].fillna(all_data['Utilities'].mode()[0])
```

```python
# Total square footage across all floors
all_data['TotalSF'] = (all_data['TotalBsmtSF'] + 
                       all_data['1stFlrSF'] + 
                       all_data['2ndFlrSF'])

# Total bathrooms (half baths count as 0.5)
all_data['TotalBaths'] = (all_data['FullBath'] + 
                          all_data['BsmtFullBath'] +
                          0.5 * (all_data['HalfBath'] + all_data['BsmtHalfBath']))

# House age and remodel age at time of sale
all_data['HouseAge'] = all_data['YrSold'] - all_data['YearBuilt']
all_data['RemodAge'] = all_data['YrSold'] - all_data['YearRemodAdd']

# Was the house remodelled at all?
all_data['WasRemodelled'] = (all_data['YearBuilt'] != all_data['YearRemodAdd']).astype(int)

# Is it a new house (sold in the same year it was built)?
all_data['IsNew'] = (all_data['YearBuilt'] == all_data['YrSold']).astype(int)

print("New features created:", ['TotalSF', 'TotalBaths', 'HouseAge', 'RemodAge', 'WasRemodelled', 'IsNew'])
```

```python
# These columns have a natural order: Poor < Fair < Good < Excellent
# We encode that order explicitly as numbers
quality_map = {'None': 0, 'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5}

quality_cols = ['ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond',
                'HeatingQC', 'KitchenQual', 'FireplaceQu',
                'GarageQual', 'GarageCond', 'PoolQC']

for col in quality_cols:
    all_data[col] = all_data[col].map(quality_map)

# Finished basement type also has an order
all_data['BsmtFinType1'] = all_data['BsmtFinType1'].map(
    {'None': 0, 'Unf': 1, 'LwQ': 2, 'Rec': 3, 'BLQ': 4, 'ALQ': 5, 'GLQ': 6})

all_data['BsmtFinType2'] = all_data['BsmtFinType2'].map(
    {'None': 0, 'Unf': 1, 'LwQ': 2, 'Rec': 3, 'BLQ': 4, 'ALQ': 5, 'GLQ': 6})

print("Ordinal encoding done.")
```

```python
# These columns have no natural order — just different categories
nominal_cols = ['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour',
                'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood',
                'Condition1', 'Condition2', 'BldgType', 'HouseStyle',
                'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd',
                'MasVnrType', 'Foundation', 'BsmtExposure',
                'Heating', 'CentralAir', 'Electrical', 'Functional',
                'GarageType', 'GarageFinish', 'PavedDrive',
                'Fence', 'MiscFeature', 'SaleType', 'SaleCondition']

all_data = pd.get_dummies(all_data, columns=nominal_cols, drop_first=True)

print(f"Total columns after encoding: {all_data.shape[1]}")
```

```python
# Separate back into train and test using the original sizes
ntrain = df.shape[0]

X_train = all_data[:ntrain].copy()
X_test  = all_data[ntrain:].copy()

print(f"X_train: {X_train.shape}")
print(f"X_test:  {X_test.shape}")
print(f"y (target): {y.shape}")
```

```python
from sklearn.preprocessing import RobustScaler

numeric_cols = X_train.select_dtypes(include=[np.number]).columns

scaler = RobustScaler()
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols]  = scaler.transform(X_test[numeric_cols])

print("Scaling complete.")
print(f"Final training shape: {X_train.shape}")
```

```python
print("=== Final check ===")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape:  {X_test.shape}")
print(f"y shape:       {y.shape}")
print(f"Any nulls in X_train: {X_train.isnull().sum().sum()}")
print(f"Any nulls in X_test:  {X_test.isnull().sum().sum()}")
print(f"\nSample of y (log-transformed prices):\n{y.describe().round(3)}")
```

# Model Training and Cross-Validation

```python
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import mean_squared_error
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
def rmse_cv(model, X, y):
    scores = cross_val_score(
        model, X, y,
        scoring='neg_root_mean_squared_error',
        cv=kf
    )
    return -scores  # sklearn returns negative scores, flip to positive

def evaluate(model, X, y, name):
    scores = rmse_cv(model, X, y)
    print(f"{name}")
    print(f"  RMSE scores: {scores.round(4)}")
    print(f"  Mean: {scores.mean():.4f}  |  Std: {scores.std():.4f}")
    print()
    return scores.mean()
```

```python
lr = LinearRegression()
lr_score = evaluate(lr, X_train, y, "Linear Regression")
```

```python
ridge = Ridge(alpha=10)
ridge_score = evaluate(ridge, X_train, y, "Ridge Regression")
```

```python
lasso = Lasso(alpha=0.0005)
lasso_score = evaluate(lasso, X_train, y, "Lasso Regression")
```

```python
gbr = GradientBoostingRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=4,
    min_samples_leaf=15,
    random_state=42
)
gbr_score = evaluate(gbr, X_train, y, "Gradient Boosting")
```

```python
results = {
    'Linear Regression': lr_score,
    'Ridge':             ridge_score,
    'Lasso':             lasso_score,
    'Gradient Boosting': gbr_score,
}

print("=== Model Comparison ===")
print(f"{'Model':<25} {'CV RMSE':>10}")
print("-" * 37)
for name, score in sorted(results.items(), key=lambda x: x[1]):
    print(f"{name:<25} {score:>10.4f}")

best_model_name = min(results, key=results.get)
print(f"\nBest model so far: {best_model_name}")
```

```python
# Fit Gradient Boosting on the full training set
gbr.fit(X_train, y)

# Get feature importances
importance = pd.Series(gbr.feature_importances_, index=X_train.columns)
top20 = importance.sort_values(ascending=False).head(20)

plt.figure(figsize=(10, 7))
top20.sort_values().plot(kind='barh', color='steelblue')
plt.title('Top 20 most important features (Gradient Boosting)')
plt.xlabel('Feature importance score')
plt.tight_layout()
plt.show()
```

```python
# Make predictions on training data to check they look sensible
train_preds = gbr.predict(X_train)

# Convert back from log scale to real dollars
actual_prices    = np.expm1(y)
predicted_prices = np.expm1(train_preds)

# Compare a sample
comparison = pd.DataFrame({
    'Actual ($)':    actual_prices.values[:10].round(0),
    'Predicted ($)': predicted_prices[:10].round(0),
    'Difference ($)': (predicted_prices[:10] - actual_prices.values[:10]).round(0)
})
print(comparison.to_string(index=False))
```
